Task - 2 - Composite Score Ranking: Tasks:

Using the SouthAsia DataFrame, create a new column called Composite Score that combines the following metrics: Composite Score = 0.40 × GDP per Capita + 0.30 × Social Support
0.30 × Healthy Life Expectancy
Rank the South Asian countries based on the Composite Score in descending order.
Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.
Discuss whether the rankings based on the Composite Score align with the original Score - support your discussion with some visualization plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("World Happinesss Report.csv")

# South Asia countries
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka", "Srilanka"
]

# Filter
south_df = df[df["Country name"].isin(south_asia)].copy()

# Columns
gdp_col = "Log GDP per capita"
support_col = "Social support"
health_col = "Healthy life expectancy"
score_col = "score"

# Create Composite Score
south_df["Composite Score"] = (
    0.40 * south_df[gdp_col] +
    0.30 * south_df[support_col] +
    0.30 * south_df[health_col]
)

# Ranking
south_df = south_df.sort_values("Composite Score", ascending=False)
print(south_df[["Country name", "Composite Score"]])


# GRAPH 1 – TOP 5 BAR CHART

top5 = south_df.head(5)

plt.figure(figsize=(9, 5))
plt.barh(top5["Country name"][::-1], top5["Composite Score"][::-1])
plt.xlabel("Composite Score")
plt.title("Top 5 South Asian Countries by Composite Score")
plt.tight_layout()
plt.show()

# GRAPH 2 – SCATTER PLOT

x = south_df["Composite Score"]
y = south_df[score_col]

plt.figure(figsize=(7, 5))
plt.scatter(x, y)

# regression line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)

plt.xlabel("Composite Score")
plt.ylabel("Original Score")
plt.title("Composite Score vs Original Happiness Score")
plt.tight_layout()
plt.show()
```

Task - 3 - Outlier Detection: Tasks:

Identify outlier countries in South Asia based on their Score and GDP per Capita.
Define outliers using the 1.5 × IQR rule.
Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import matplotlib.pyplot as plt

# 1. Calculate IQR for Score and GDP per Capita
Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# 2. Define outliers using 1.5 * IQR rule
score_lower = Q1_score - 1.5 * IQR_score
score_upper = Q3_score + 1.5 * IQR_score

gdp_lower = Q1_gdp - 1.5 * IQR_gdp
gdp_upper = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers
outliers = south_asia_df[
    (south_asia_df['score'] < score_lower) | (south_asia_df['score'] > score_upper) |
    (south_asia_df['Log GDP per capita'] < gdp_lower) | (south_asia_df['Log GDP per capita'] > gdp_upper)
]

print("Outlier countries in South Asia:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])

# 3. Scatter plot highlighting outliers
plt.figure(figsize=(8,6))
# Normal points
normal = south_asia_df[~south_asia_df.index.isin(outliers.index)]
plt.scatter(normal['Log GDP per capita'], normal['score'], color='blue', label='Normal', s=100)
# Outliers
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outlier', s=150)

# Annotate outliers
for i, txt in enumerate(outliers['Country name']):
    plt.annotate(txt, (outliers['Log GDP per capita'].iloc[i], outliers['score'].iloc[i]), color='red')

plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("South Asia: Score vs GDP per Capita with Outliers Highlighted")
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("World Happinesss Report.csv")  # replace with your file name

# South Asia countries
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives",
                        "Nepal", "Pakistan", "Sri Lanka"]

# Middle East countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Create DataFrames
south_asia_df = df[df['Country name'].isin(south_asia_countries)]
middle_east_df = df[df['Country name'].isin(middle_east_countries)]

# Descriptive statistics: mean and standard deviation of Happiness Score
south_asia_mean = south_asia_df['score'].mean()
south_asia_std = south_asia_df['score'].std()

middle_east_mean = middle_east_df['score'].mean()
middle_east_std = middle_east_df['score'].std()

# Print results
print("South Asia - Mean Score:", round(south_asia_mean, 2), ", Std Dev:", round(south_asia_std, 2))
print("Middle East - Mean Score:", round(middle_east_mean, 2), ", Std Dev:", round(middle_east_std, 2))

# Compare which region is happier on average
if south_asia_mean > middle_east_mean:
    print("South Asia has higher average happiness scores.")
elif middle_east_mean > south_asia_mean:
    print("Middle East has higher average happiness scores.")
else:
    print("Both regions have the same average happiness scores.")
```

6. Outlier Detection: • Identify outlier countries in both regions based on Score and GDP per Capita. • Plot these outliers and discuss their implications.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Function to detect outliers using 1.5 * IQR rule
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers in South Asia
outliers_sa_score = detect_outliers(south_asia_df, 'score')
outliers_sa_gdp = detect_outliers(south_asia_df, 'Log GDP per capita')

# Detect outliers in Middle East
outliers_me_score = detect_outliers(middle_east_df, 'score')
outliers_me_gdp = detect_outliers(middle_east_df, 'Log GDP per capita')

# Combine Score and GDP outliers for plotting
outliers_sa = pd.concat([outliers_sa_score, outliers_sa_gdp]).drop_duplicates()
outliers_me = pd.concat([outliers_me_score, outliers_me_gdp]).drop_duplicates()

# Scatter plot for South Asia
plt.figure(figsize=(8,6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='Countries')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers_sa, color='red', s=100, label='Outliers')
plt.title("South Asia: Score vs GDP per Capita (Outliers in Red)")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()

# Scatter plot for Middle East
plt.figure(figsize=(8,6))
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_df, label='Countries')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers_me, color='red', s=100, label='Outliers')
plt.title("Middle East: Score vs GDP per Capita (Outliers in Red)")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()

# Display outlier countries
print("South Asia Outliers:\n", outliers_sa[['Country name', 'score', 'Log GDP per capita']], "\n")
print("Middle East Outliers:\n", outliers_me[['Country name', 'score', 'Log GDP per capita']], "\n")
```
